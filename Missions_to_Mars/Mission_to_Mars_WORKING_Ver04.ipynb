```python
#Dependencies
import requests
import os
from bs4 import BeautifulSoup as bs
import pymongo
from splinter import Browser
import pandas as pd
import time
```

```python
executable_path = {"executable_path": "/___Rice/HomeWork/Home_Work_12_Web_Scraping/WIP/2/chromedriver"}
browser = Browser("chrome", **executable_path, headless=False)
```

```python
# Scraping 1: NASA Mars News
url='https://mars.nasa.gov/news/'
browser.visit(url)
```

```python
# To assure that all data sent by the website gets captured 
time.sleep(5)
```

```python
html=browser.html
```

```python
soup=bs(html, 'html.parser')
```

```python
# Saving the HTML data to a file for future reference or possible use
with open('Mars_Mars.txt', 'w', encoding='utf-8') as f_out:
    f_out.write(soup.prettify())
```

```python
# Verifying data sufficiency
adiv = soup.find_all('div')
adiv_text = soup.find_all( class_="list_text")
adiv_teaser=soup.find_all( class_="article_teaser_body")
adiv_content=soup.find_all( class_="content_title")
print(len(adiv), len(adiv_text), len(adiv_teaser), len(adiv_content) )
```

```python
# Closing the browser so as NOT to overload Windows System
browser.quit()
```

```python
# Iterating over ALL divs and capturing ONLY those items where all 3 elements: Heading, Teaser, & Body exist
items=[]
xcount=0
for ad in adiv:
    lxt=ad.find('div', class_='list_text')
    tsr=ad.find('div', class_='article_teaser_body')
    ctx=ad.find('div', class_='content_title')
    clxt=str(type(lxt))
    ctsr=str(type(tsr))
    cctx=str(type(ctx))
    print(clxt, ctsr, cctx)
    if 'element' in clxt and 'element' in ctsr and 'element' in cctx:
        xcount=xcount+1
        txt_ctx=ctx.text
        txt_lxt=lxt.text
        txt_tsr=tsr.text
        print("\n\n")
        print(txt_ctx)
        print(txt_lxt)
        print(txt_tsr)
        print("\n")
        items.append((txt_ctx, txt_lxt, txt_tsr))

print(f"\nAll Three in: {xcount}")
```

```python
# Preparing DataFrame and saving to a CSV file
df = pd.DataFrame(items, columns=['Title', 'Heading_Text', 'Teaser'])
df.to_csv('Mars_Headings.csv', index=False, encoding='utf-8')
df.tail()
```
